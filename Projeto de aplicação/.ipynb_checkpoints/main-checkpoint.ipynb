```python
import pandas as pd
from sklearn import tree
```

```python
# importando os dados e verificando as colunas
dataset = pd.read_csv('./autism_screening.csv')
dataset.head()
```

```python
# removendo algumas colunas irrelevantes e renomeando outras para melhorar a legibilidade
dataset = dataset.drop(columns =['ethnicity', 'contry_of_res', 'used_app_before', 'age_desc', 'relation'])
dataset = dataset.rename(columns={'austim': 'autism_history', 'Class/ASD': 'autism'})
dataset.head()
```

```python
# fazendo hot encoding nas colunas de dados categóricos
dataset = pd.get_dummies(dataset, columns=['autism_history', 'gender', 'jundice', 'autism'])
dataset.head()
```
